# *Importing Libraries and Packages*

```python
df.drop(['id'],axis=1,inplace=True)
df.drop(['Unnamed: 32'],axis=1,inplace=True)
df
```

```python
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df
```

```python
X=df.iloc[:,1:]
X
```

```python
y=df.iloc[:,0]
y
```

# *Scaling*

```python
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X
```

# *Train Test Split*

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.3)
```

# *Model Creation*

```python
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
lr=LogisticRegression()
models=[knn,dt,rf,lr]
accuracy=[]
for model in models:
  print('*******************',model,'*******************')
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print(classification_report(y_test,y_pred))
  accuracy.append(accuracy_score(y_test,y_pred)*100)
```

```python
print(accuracy)
```

```python
algo=['knn','dt','rf','lr']
plt.bar(algo,accuracy)
plt.plot()
plt.xlabel('Algorithms used')
plt.ylabel('Accuracy_score')
plt.title('Algorithms used v/s Accuracyscore')
plt.show()
```

# *OverSampling*

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=15)
XO,yO=ros.fit_resample(X,y)
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', p=2)
knn_classifier.fit(XO,yO)
knn_y_pred=knn_classifier.predict(X_test)
print(classification_report(y_test,knn_y_pred))
```

```python
print('Accuracy score',accuracy_score(y_test,knn_y_pred)*100)
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,knn_y_pred))
```
